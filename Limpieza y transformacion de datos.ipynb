```python
#importamos las librerias que se van a usar
import pandas as pd

# Leer un archivo Excel y seleccionar una hoja en específico
df = pd.read_excel("Encuesta.xlsx", sheet_name="Hoja1")

# Mostrar las primeras filas
print(df.head())
```

```python
# Cantidad de valores faltantes por columna
print("*****************Registros**********************")
print("Cantidad de valores faltantes por columna:")
print(df.isnull().sum())
```

```python
#Renombro los nombres de las columnas
df = df.rename(columns={
    "DEPARTAMENTO DONDE NACIO": "Departamento",
    "MUNICIPIO DONDE NACIO": "Municipio",
    "ESCOLARIDAD \n(Indicar el último grado del que tiene diploma)": "Escolaridad",
    "SI SABE INGLÉS \nEN QUÉ NIVEL CONSIDERA DOMINA INGLÉS (Básico, intermedio, avanzado)":"Nivel de Ingles",
    "PROFESION / OFICIO EN EEUU":"OFICIO",
    "TIEMPO DE EXPERIENCIA EN SU  OFICIO EN EEUU":"Año Experiencia",
    "ESTADO EN EL QUE ESTUVO EN EEUU":"ESTADO USA",
    "TIENE FAMILIA EN EEUU":"Familiares USA"
})

# Eliminar filas de la columnas Que tengan valores NA o vacios
df = df.dropna(subset=["EDAD"])
df = df.dropna(subset=["Departamento"])
# Convertir la columna "EDAD" a entero
df["EDAD"] = df["EDAD"].astype(int)
print("*****************Registros**********************")
print("Cantidad de valores faltantes por columna:")
print(df.isnull().sum())
```

```python
# Relleno valores vacíos en municipio con "Desconocido"
df["Municipio"] = df["Municipio"].fillna("Desconocido")

print(df["Municipio"].value_counts().head())
```

```python
df = df.drop(columns=["DEPARTAMENTO DONDE RESIDIRA"])
df = df.drop(columns=["LE GUSTARÍA APOYO PARA EMPLEARSE EN GUATEMALA"])
```

```python
print(df.isnull().sum())
```

```python
# Seleccionar las columnas categoricas
cols_categoricas = df.select_dtypes(include=['object', 'category']).columns

# Revisar los valores únicos por cada columna
for col in cols_categoricas:
    print(f"\nColumna: {col}")
    print(df[col].unique())
```

```python
#normalizar columna de Escolaridad se va agrupar todo en categorías resumidas para que tengan sentido analítico
def categorizar_escolaridad(valor):
    if pd.isna(valor):
        return "No indicó"
    v = valor.strip().lower()

    # 1. Sin estudios
    if any(x in v for x in ["ningun", "no tiene", "no estudio", "analfab", "no leer"]):
        return "Sin estudios"

    # 2. Primaria
    if "primaria" in v or "alfabetización" in v or "1 primaria" in v or "2 primaria" in v or "3 primaria" in v or "4 primaria" in v or "5 primaria" in v or "6 primaria" in v:
        return "Primaria"

    # 3. Básicos
    if "basico" in v or "básico" in v or "secundaria" in v:
        return "Básicos"

    # 4. Diversificado
    if any(x in v for x in ["bachiller", "perito", "magisterio", "diversificado", "secretaria", "auxiliar"]):
        return "Diversificado"

    # 5. Universitario
    if "universidad" in v or "universitario" in v or "ingenier" in v or "licenciatura" in v or "profesorado" in v:
        return "Universitario"

    # 6. Extranjero
    if "high school" in v or "eeuu" in v or "estados unidos" in v:
        return "Estudios en el extranjero"

    # 7. Otros
    return "Otros"

# Aplicar al DataFrame
df["Escolaridad"] = df["Escolaridad"].apply(categorizar_escolaridad)

print(df["Escolaridad"].value_counts())
```

```python
#normalizar columna de Escolaridad se va agrupar todo en categorías resumidas para que tengan sentido analítico
def normalizar_nivel(valor):
    if pd.isna(valor):
        return "Sin estudios"
    v = valor.strip().lower()

    # Sin estudios / Analfabetismo
    if "analfabetismo" in v or "no tiene" in v:
        return "Sin estudios"

    # Primaria
    if "primaria" in v:
        return "Primaria"

    # Secundaria / Básicos
    if "secundaria" in v or "basico" in v or "basicos" in v:
        return "Secundaria / Básicos"

    # Diversificado
    if "diversificado" in v:
        return "Diversificado"

    # Universitario
    if "universitario" in v:
        return "Universitario"

    # Otros / No indicó
    return "Otros"

# Aplicar la función al DataFrame
df["NIVEL ESCOLARIDAD"] = df["NIVEL ESCOLARIDAD"].apply(normalizar_nivel)

# Revisar conteo por categoría
print(df["NIVEL ESCOLARIDAD"].value_counts())
```

```python
# Normalizar columna de ingles
df["SABE INGLES (SI/NO)"] = df["SABE INGLES (SI/NO)"].replace({
    "Sí": "Si",
    "Si": "Si",
    "si": "Si",
    "NO": "No",
    "no": "No",
    "No": "No",
    "Un poco": "Si"
})

# Rellenar los valores faltantes con "No"
df["SABE INGLES (SI/NO)"] = df["SABE INGLES (SI/NO)"].fillna("No")

print(df["SABE INGLES (SI/NO)"].value_counts())
```

```python
import numpy as np
# Normalizar valores
df["Nivel de Ingles"] = df["Nivel de Ingles"].replace({
    "Basico": "Basico",
    "Basico ": "Basico",
    "basico": "Basico",
    "Básico": "Basico",
    "Leer":"Basico",
    "Escribir":"Basico",
    
    "Intermedio": "Intermedio",
    "Inglés intermedio": "Intermedio",
    "Hablar": "Intermedio",
    "Leer, Hablar":"Intermedio",
    "Escribir, Hablar":"Intermedio",
    "Leer, Escribir":"Intermedio",
    
    "Avanzado": "Avanzado",
    "avanzado": "Avanzado",
    "Leer, Escribir, Hablar":"Avanzado",
})

# Convertir a NaN todo lo que no esté en las tres categorías
df["Nivel de Ingles"] = df["Nivel de Ingles"].where(df["Nivel de Ingles"].isin(["Basico", "Intermedio", "Avanzado"]), np.nan)

print(df["Nivel de Ingles"].value_counts(dropna=False))
```

```python
import re
#Normalizar la columna experiencia por solo numeros, en el caso de meses se maneja de forma decimal 
#En caso que solo tenga el oficio se asignara el valor nan
def normalizar_experiencia(value):
    if pd.isna(value):
        return np.nan
    
    text = str(value).lower().strip()
    
    # Caso: no tiene experiencia
    if "no tiene" in text:
        return 0.0
    
    total_years = 0.0
    
    # Busca años como "4 años", "15 AÑOS", "2.5 años"
    years = re.findall(r"(\d+(?:\.\d+)?)\s*años?", text)
    if years:
        total_years += sum(float(y) for y in years)
    
    # Buscar los meses
    months = re.findall(r"(\d+(?:\.\d+)?)\s*mes", text)
    if months:
        total_years += sum(float(m) for m in months) / 12.0
    
    # Detectar "medio" o "y medio" -> +0.5 año
    if re.search(r"\bmedio\b", text):
        total_years += 0.5
    
    # Si no se encontró nada numérico -> NaN (probablemente solo oficio)
    if total_years == 0.0:
        return np.nan
    
    return round(total_years, 1)

#  Aplica la función en la columna
df["Año Experiencia"] = df["Año Experiencia"].apply(normalizar_experiencia)

# Ejemplo de conteo
print(df["Año Experiencia"].value_counts(dropna=False).head(20))
```
